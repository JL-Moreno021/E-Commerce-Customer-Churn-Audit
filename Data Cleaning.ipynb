# Data Cleaning

```python
import pandas as pd
df = pd.read_csv("customer churn.csv")
```

## Data Observations Checking
<hr>


**_Numeric Dtypes_**<br>
>- Invalid numeric values
>- Duplicate values on unique field

**_String Dtypes_**
>- Duplicate

<hr>

### Numeric Datatypes

Use `describe()` to view invalid observations

```python
df.describe()
```

**_Findings_**
>---
**_Age:_**<br>
    - have min value of `-10`
    - have a big outlier

**_Customer Service Calls_**<br>
    - have min value of `-1`

**_Monthly_Spend_**<br>
    - have min value of `-10`
    - have a big outlier

<hr>

**_Process_**<br>
>---
Convert all non-positive values in `Age`, `Customer Service Calls`, and `Monthly Spend` into `nan`

```python
df['Age'] = df['Age'].where(df['Age'] > 0 )
df['Customer_Service_Calls'] = df['Customer_Service_Calls'].where(df['Customer_Service_Calls'] > 0 )
df['Monthly_Spend'] = df['Monthly_Spend'].where(df['Monthly_Spend'] > 0 )

for cols in ['Age', 'Customer_Service_Calls', 'Monthly_Spend']:
    print(sorted(df[cols]))
```

```python
inv_age = df[df['Age'] > 70]
inv_age
```

```python
max_spend = df[df['Monthly_Spend'] > 290]
max_spend
```

### String Datatypes

Use `set()` to view unique observations and to identify typographical errors

```python
genders = set(df['Gender'])
regions = set(df['Region'])
subs = set(df['Subscription_Plan'])
payment = set(df['Payment_Method'])

print(genders)
print(regions)
print(subs)
print(payment)
```

**_Findings:_**
>---
**_Region_**<br>
    - Some values are not capitalized

**_Subscription Plan_**<br>
    - Some values are in uppercase
<hr>

**_Process_**
>---
Convert all to `Capitalized` format

```python
for col in ['Region','Subscription_Plan']:
    df[col] = (
        df[col]
        .str.lower()
        .str.capitalize()
        .str.strip()
    )
```
